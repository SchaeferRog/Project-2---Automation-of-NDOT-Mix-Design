# Part 2 - Cement Calculation Functions


## 1. Function Q: Water Weight (lb/yd³)
**Code Cell:** define `weight_water_Q`

**Inputs:**
- cement_A, fly_ash_B, silica_fume_C, other_SCM_D, wc_ratio_E

## 2. Function R: Volume of Cement (yd³)
**Code Cell:** define `vol_cement_R`

**Inputs:**
- cement_A, cement_sg_J

## 3. Function S: Volume of Fly Ash (yd³)
**Code Cell:** define `vol_fly_ash_S`

**Inputs:**
- fly_ash_B, fly_ash_sg_

## 4. Function T: Volume of Sulica Fume (yd³)
**Code Cell:** define `vol_silica_fume_T`

**Inputs:**
- silica_fume_C, silica_fume_sg_L

## 5. Function U: Volume of other SCM (yd³)
**Code Cell:** define `vol_other_SCM_U`

**Inputs:**
- other_SCM_D, other_SCM_sg_M

## 6. Function V: Volume of Air (yd³)
**Code Cell:** define `vol_air_V`

**Inputs:**
- air_content_F


## 7. Function W: Volume of Water as a function of Water Weight (Q) (yd³)
**Code Cell:** define `vol_water_W`

**Inputs:**
- Q

## 8. Function X: Volume of Total Aggregate (yd³)
**Code Cell:** define `vol_total_aggregate_X`

**Inputs:**
- vol_cement_R,vol_fly_ash_S,vol_silica_fume_T,vol_other_SCM_U,vol_air_V,vol_water_W

## 9. Function Y: Weight of Fine Aggregate (lb/yd³)
**Code Cell:** define `weight_fine_aggregate_Y`

**Inputs:**
- fine_aggregate_G, fine_aggregate_sg_N, vol_total_aggregate_X

## 10. Function Z: Weight of Coarse Aggregate (lb/yd³)
**Code Cell:** define `weight_coarse_aggregate_Z`

**Inputs:**
- coarse_aggregate_H, coarse_aggregate_sg_O, vol_total_aggregate_X

## 11. Function AA: Weight of Other Aggregates (lb/yd³)
**Code Cell:** define `weight_other_aggregate_AA`

**Inputs:**
- other_aggregate_I, other_aggregate_P, vol_total_aggregate_X


---

```python
### REPRESENTING EXCEL EQUATIONS IN PYTHON ###

# Function to calculate water weight
def weight_water_Q(cement_A, fly_ash_B, silica_fume_C, other_SCM_D, wc_ratio_E): 
    return (cement_A+fly_ash_B+silica_fume_C+other_SCM_D)*(wc_ratio_E)

# Function to calculate the volume of cement (R)
def vol_cement_R(cement_A, cement_sg_J):
    return cement_A/(cement_sg_J*62.4)

# Function to calculate fly ash volume
def vol_fly_ash_S(fly_ash_B, fly_ash_sg_K):
    return fly_ash_B/(fly_ash_sg_K*62.4)

# Function to calculate silica fume volume
def vol_silica_fume_T(silica_fume_C, silica_fume_sg_L):
    return silica_fume_C/(silica_fume_sg_L*62.4)

# Function to calculate other SCM volume
def vol_other_SCM_U(other_SCM_D, other_SCM_sg_M):
    return other_SCM_D/(other_SCM_sg_M*62.4)

# Define a function to calculate other air volume
def vol_air_V(air_content_F):
    return (air_content_F/100)*27

# Function to calcualte water volume (W) as a function of water weight (Q)
def vol_water_W(Q):
    return Q/62.4

# Function to calculate total aggregate volume (X)
def vol_total_aggregate_X(vol_cement_R,vol_fly_ash_S,vol_silica_fume_T,vol_other_SCM_U,vol_air_V,vol_water_W):
    return 27-vol_cement_R-vol_fly_ash_S-vol_silica_fume_T-vol_other_SCM_U-vol_air_V-vol_water_W

# Function to calculate fine aggregate weight (Y)
def weight_fine_aggregate_Y(fine_aggregate_G, fine_aggregate_sg_N, vol_total_aggregate_X):
    return (62.4*(fine_aggregate_G/100)*fine_aggregate_sg_N*vol_total_aggregate_X)

# Function for coarse aggregate weight (Z)
def weight_coarse_aggregate_Z(coarse_aggregate_H, coarse_aggregate_sg_O, vol_total_aggregate_X):
    return (62.4*(coarse_aggregate_H/100)*coarse_aggregate_sg_O*vol_total_aggregate_X)

# Function for other aggregate weight (AA)
def weight_other_aggregate_AA(other_aggregate_I, other_aggregate_P, vol_total_aggregate_X):
    return (62.4*(other_aggregate_I/100)*other_aggregate_P*vol_total_aggregate_X)
    
```

# Part 3 - Sequential User Inputs 


---

```python
### COLLECT USER INPUTS SEQUENTIALLY WITH INPUT() ###

# General info
project_no = int(input("Enter Project Number: "))
concrete_class = input("Enter Class of Concrete: ")
control_no = int(input("Enter Control Number: "))

# Cementitious material inputs
cement_A = float(input("Weight of Cement per yd3: "))
fly_ash_B = float(input("Weight of Fly Ash per yd3: "))
silica_fume_C = float(input("Weight of Silica Fume per yd3: "))
other_SCM_D = float(input("Weight of other SCM per yd3: "))

# Design parameters
wc_ratio_E = float(input("Target Water/Cement Ratio: "))
air_content_F = float(input("Target % Air Content: "))

# Aggregate proportions
fine_aggregate_G = float(input("Target % Fine Aggregate: "))
coarse_aggregate_H = float(input("Target % Coarse Aggregate: "))
other_aggregate_I = float(input("Target % Other Aggregate: "))


# Specific gravities
cement_sg_J = float(input("Weight of Cement per yd3 Specific Gravity: "))
fly_ash_sg_K = float(input("Weight of Fly Ash per yd3 Specific Gravity: "))
silica_fume_sg_L = float(input("Weight of Silica Fume per yd3 Specific Gravity: "))
other_SCM_sg_M = float(input("Weight of other SCM per yd3 Specific Gravity: "))
fine_aggregate_sg_N = float(input("Target % Fine Aggregate Specific Gravity: "))
coarse_aggregate_sg_O = float(input("Target % Coarse Aggregate Specific Gravity: "))
other_aggregate_sg_P = float(input("Target % Other Aggregate Specific Gravity: "))

# A Print message to confirm all inputs have been collected
print("\nAll user inputs collected successfully.")
```

# Part 4 - Running Completed Mix Design and Presenting Outputs

---

```python
### APPLY USER INPUTS TO FUNCTIONS ### 

# Step 1: Water weight (Q)
Q = weight_water_Q(cement_A, fly_ash_B, silica_fume_C, other_SCM_D, wc_ratio_E)

# Step 2: Cementitious volumes (R, S, T, U)
R = vol_cement_R(cement_A, cement_sg_J)
S = vol_fly_ash_S(fly_ash_B, fly_ash_sg_K)
T = vol_silica_fume_T(silica_fume_C, silica_fume_sg_L)
U = vol_other_SCM_U(other_SCM_D, other_SCM_sg_M)

# Step 3: Air and water volumes (V, W)
V = vol_air_V(air_content_F)
W = vol_water_W(Q)

# Step 4: Total aggregate volume (X)
X = vol_total_aggregate_X(R,S,T,U,V,W)

# Step 5: Aggregate weights (Y, Z, AA)
Y = weight_fine_aggregate_Y(fine_aggregate_G, fine_aggregate_sg_N, X)
Z = weight_coarse_aggregate_Z(coarse_aggregate_H,coarse_aggregate_sg_O,X)
AA = weight_other_aggregate_AA(other_aggregate_I, other_aggregate_sg_P, X)

### REPORT OUTPUTS ###

# Project and Mix Details 
print("\n---------------------------------------------")
print(f"PROJECT NUMBER: {project_no}")
print(f"CONCRETE CLASS: {concrete_class}")
print(f"CONTROL NUMBER: {control_no}")
print("\n---------------------------------------------")

# General 
print(" NDOT Concrete Mix Design – Weight Summary")
print("         (1 Cubic Yard of Concrete)")
print("\n---------------------------------------------")


# Cementitious Material Outputs
print(f"CEMENT: {cement_A:.2f} lb")
print(f"FLY ASH: {fly_ash_B:.2f} lb")
print(f"SILICA FUME: {silica_fume_C:.2f} lb")
print(f"OTHER SCM: {other_SCM_D:.2f} lb")

# Aggregate Outputs
print(f"FINE AGGREGATE: {Y:.2f} lb")
print(f"COARSE AGGREGATE: {Z:.2f} lb")
print(f"OTHER AGGREGATE: {AA:.2f} lb")

# Water Output
print(f"WATER: {Q:.2f} lb")
print("---------------------------------------------")

# A Note
print("Weight Summary Complete!")
```
